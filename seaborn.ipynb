```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tips =pd.read_csv('/content/tips.csv')
df_tips.head()
```

# **without download dataset**

```python
df_tips =sns.load_dataset('tips')
df_tips.head()
```

# scatter plot

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='total_bill', y='tip', data=df_tips, hue='sex',style='sex')
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=df_tips,col= 'smoker',row='sex',size='size',hue='size', style='smoker')
plt.show()
```

```python
sns.countplot(x='sex',data=df_tips,hue='smoker')
```

```python
sns.boxplot(y='tip',data=df_tips, x='sex', hue='smoker')
```

```python
sns.boxplot(x='tip',data=df_tips)
```

```python
sns.catplot(x='total_bill', y='tip', data=df_tips,col= 'smoker',row='sex')
plt.show()
```

```python
sns.catplot(x='sex', y='tip', data=df_tips,col= 'smoker',kind='box')
plt.show()
```

# **histogram**

```python
plt.figure(figsize=(10,6))
sns.histplot(x='total_bill',data=df_tips,bins=25,hue='smoker')
plt.title('total bill in dataset')
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(20,10))
plt.sca(ax[0])
sns.histplot(x='total_bill',data=df_tips[df_tips['smoker']=='No'],bins=25)
plt.title('smoker is no')
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.sca(ax[1])
sns.histplot(x='total_bill',data=df_tips[df_tips['smoker']=='Yes'],bins=25)
plt.title('smoker is yes')
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x='total_bill',data=df_tips,bins=25,hue='smoker',kde=True)
plt.title('total bill in dataset')
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.show()
```
