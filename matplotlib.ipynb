```python
import matplotlib.pyplot as plt
import numpy as np
area =np.array ([10,20,30,40,50])
price = np.array([100,200,300,400,500])
price2 = np.array([80,220,150,420,520])
plt.title("Heba")
plt.xlabel("area")
plt.ylabel("price")
#or a = plt.axes()
#a.set(title="heba",xlabel="area",ylabel="price")
#plt.xlim(20, 40)
#plt.ylim(200, 405)
plt.plot(area,price , '-b', label = "Tower 1")
plt.plot(area,price2 , ':r', label = "Tower 2")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_house = pd.read_csv('/content/housing (1).csv')
```

```python
df_house.head()
```

## *scatter plot*

```python
x= np.linspace(10,50,60)
y= x**2
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_house['longitude'],df_house['latitude'],alpha=0.4,marker= "+", s=50)
plt.xlabel('longitude', color ='red')
plt.ylabel('latitude', color ='red')
plt.title('longitude vs latitude', fontsize=18, color ='red')

plt.show()
```

```python
plt.figure(figsize=(9,6))
sc= plt.scatter(df_house['longitude'],df_house['latitude'],alpha=0.4,marker= "o",
            s=df_house['population']/100, c=df_house['median_house_value'],cmap=plt.get_cmap('jet'),
            label= 'population')
plt.colorbar(sc)
plt.xlabel('longitude', color ='red')
plt.ylabel('latitude', color ='red')
plt.title('longitude vs latitude', fontsize=18, color ='red')
plt.legend()
plt.show()
```

```python
df_house.plot(x='longitude',y= 'latitude',kind='scatter',figsize=(9,6),
             s= df_house['population']/100, c='median_house_value' ,cmap=plt.get_cmap('jet'))
```

```python
plt.figure(figsize=(9,6))
x= 2*np.random.rand(100,1)
y= 4+3*x+np.random.rand(100,1)
z= 4-3*x+np.random.rand(100,1)
plt.scatter(x,y,color='r',label='x&y')
plt.scatter(x,z,color='g',label='x&z')
plt.title('scatterplot for some test data', fontsize=12)
plt.xlabel('x',fontsize=12)
plt.legend()
plt.show()
```

```python
x= np.linspace(1,5,100)
y =np.exp(x)
z= np.log2(x)
plt.plot(x,y,'r +', markersize=10)
```

```python
plt.plot(x,y,'r')
plt.plot(x,z,'k')
plt.ylim(0,20)
plt.show()
```

```python
plt.plot(x,z)
```

```python
fig, ax=plt.subplots()
ax.plot(x,y,color='blue',label='exp(x)')
ax.set_xlabel('x',fontsize=14)
ax.set_ylabel('y', color='blue',fontsize=14)
ax2=ax.twinx()
ax2.plot(x,z,color='red',label='log10(x)')
ax2.set_ylabel('z',color='red',fontsize=14)
plt.title('x vs y & x vs z', fontsize=14)
plt.show()
```
