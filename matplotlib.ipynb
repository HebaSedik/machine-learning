```python
###ارسم بنقط
import matplotlib.pyplot as plt
#plt.style.use("classic")
#plt.style.use("bmh")  ###الخلفيه مربعات ولونها رمادي
#plt.style.use("seaborn") ###الوان مائيه
#plt.style.use("seaborn-whitegrid")
#plt.style.use("seaborn-paper")
plt.style.use("dark_background")
x = [10,20,30,40,50]
y = [100,200,300,400,500]
y2 = [120,220,320,420,520]
plt.plot(x,y, color ="red", linewidth =3, linestyle="--")   ###يعمل خط
plt.plot(x,y2, color ="blue", linewidth =3, linestyle="-.")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
area =np.array ([10,20,30,40,50])
price = np.array([100,200,300,400,500])
price2 = np.array([80,220,150,420,520])
plt.title("Heba")
plt.xlabel("area")
plt.ylabel("price")
#or a = plt.axes()
#a.set(title="heba",xlabel="area",ylabel="price")
#plt.xlim(20, 40)
#plt.ylim(200, 405)
plt.plot(area,price , '-b', label = "Tower 1")
plt.plot(area,price2 , ':r', label = "Tower 2")
plt.legend()
plt.show()
```

```python
### افصل الاشكال
import matplotlib.pyplot as plt
import numpy as np
area =np.array ([10,20,30,40,50])
price = np.array([100,200,300,400,500])
price2 = np.array([80,220,150,420,520])
plt.subplot(2,1,1)  # row #col #num
plt.plot(area , price)
plt.subplot(2,1,2)
plt.plot(area , price2)
plt.show()
```

```python
##عدد التكرار
import matplotlib.pyplot as plt
import numpy as np
data = np.array(['heba', 'mohamed' ,'sedik','heba tallah', 'heba','sedik'])
plt.hist(data , histtype = 'bar',alpha=0.6, color = 'red',bins=7)
#'bar' ,'barstracked' , step' , stepfilled'
plt.show()
```

```python
##لو عندي اكتر من مصفوفه ###احط عدد الوان بعدد العناصر ف المصفوفه الواحده
import matplotlib.pyplot as plt
import numpy as np
data = np.array([['heba', 'mohamed'] ,['sedik','heba tallah'], ['heba','sedik']])
plt.hist(data , histtype = 'bar',alpha=0.6, color = ['red','green'],bins=7)
#'bar' ,'barstracked' , step' , stepfilled'
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['heba', 'mohamed' , 'sedik'])
y= np.array([3,2,5])
plt.bar(x,y,color= 'red',alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
name =['heba', 'mohamed' , 'sedik']
cars=[3,2,4]
plt.pie(cars ,labels= name ,labeldistance =1.1,explode = [0.2,0,0] # اقطع جزء معين
        ,radius=1.1, startangle =0, colors=['red','green','blue']
        ,rotatelabels=False, autopct ='%1.0f%%')
plt.show()
```

```python
##ارسم النقط 3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
ax = plt.axes(projection ='3d')
x = np.array ([100,200,300,400,500])
y = np.array ([1000,1500,2000,1800,2500])
z = np.array ([2,2,3,4,4])
ax.plot3D (x,y,z,color='black')
ax.scatter3D(x,y,z,color='black')
plt.show()
```
