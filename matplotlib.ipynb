```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= np.linspace(10,50,60)
y= x**2
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_house['longitude'],df_house['latitude'],alpha=0.4,marker= "+", s=50)
plt.xlabel('longitude', color ='red')
plt.ylabel('latitude', color ='red')
plt.title('longitude vs latitude', fontsize=18, color ='red')

plt.show()
```

```python
plt.figure(figsize=(9,6))
sc= plt.scatter(df_house['longitude'],df_house['latitude'],alpha=0.4,marker= "o",
            s=df_house['population']/100, c=df_house['median_house_value'],cmap=plt.get_cmap('jet'),
            label= 'population')
plt.colorbar(sc)
plt.xlabel('longitude', color ='red')
plt.ylabel('latitude', color ='red')
plt.title('longitude vs latitude', fontsize=18, color ='red')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,6))
x= 2*np.random.rand(100,1)
y= 4+3*x+np.random.rand(100,1)
z= 4-3*x+np.random.rand(100,1)
plt.scatter(x,y,color='r',label='x&y')
plt.scatter(x,z,color='g',label='x&z')
plt.title('scatterplot for some test data', fontsize=12)
plt.xlabel('x',fontsize=12)
plt.legend()
plt.show()
```

```python
plt.plot(x,y,'r')
plt.plot(x,z,'k')
plt.ylim(0,20)
plt.show()
```

```python
plt.plot(x,z)
```

```python
fig, ax=plt.subplots()
ax.plot(x,y,color='blue',label='exp(x)')
ax.set_xlabel('x',fontsize=14)
ax.set_ylabel('y', color='blue',fontsize=14)
plt.legend()
ax2=ax.twinx()
ax2.plot(x,z,color='red',label='log10(x)')
ax2.set_ylabel('z',color='red',fontsize=14)
plt.title('x vs y & x vs z', fontsize=14)
plt.legend(loc='upper right')
plt.show()
```
